## Введение

<p>Работа со спутниковыми данными сопряжена с рядом проблем: большое разнообразие форматов данных, их источников, объем данных, различие способов их анализа и визуализации. В то же время постоянно исследуются пути решения этих проблем, результатом чего стало появление таких стандартов, как <a href="http://www.unidata.ucar.edu/software/netcdf/index.html">NetCDF</a> и HDF, а также ряда конвенций, определяющих структуру данных, в частности <a href="http://cf-pcmdi.llnl.gov/documents/cf-conventions/latest-cf-conventions-document-1">CF</a>. Данные других форматов, как то бинарные, csv или ascii, также стали легкодоступными благодаря протоколу DAP и построенному на его основе серверу <a href="http://opendap.org">OPeNDAP</a>. Такой сервер делает прозрачной работу с метеорологическими, климатическими, океанографическими и другими данными, предоставляя единый интерфейс доступа.</p>

<p>
<a href="http://opendap.solab.rshu.ru:8080/opendap/">OPeNDAP-сервер</a> лаборатории предоставляет доступ к ряду продуктов, доступных также через более дружественный интерфейс веб-каталога <a href="http://satin.rshu.ru">SATIN</a>. В качестве клиента может выступать любой из множества инструментов для работы с данными о Земле: Matlab, IDV, Panoply, Ferret, Pydap.
</p>
